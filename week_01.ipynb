#Given a list, print True if the sum of all the elements are greater than product of them.

```python
def sum_greater_than_product(lst):
    total_sum = sum(lst)

    total_product = 1
    for num in lst:
        total_product *= num

    return total_sum > total_product

lst = [1, 0, 0]
print(sum_greater_than_product(lst))
```

```python
def is_palindrome(num):
    if num < 0:
        return False

    temp, rev = num, 0
    while temp > 0:
        digit = temp % 10
        rev = rev * 10 + digit
        temp //= 10

    return num == rev

print(is_palindrome(121))
```

```python
def points_of_team(wins, losses, draws):
    return wins * 3 + draws

points_of_team(5, 1, 2)
```

```python
lst = [121, 133, 232, 1221, 1334331]

def sum_of_palindrome(lst):
    sum_pal = 0
    for i in lst:
        if is_palindrome(i):
            sum_pal += i
    return sum_pal

print(sum_of_palindrome(lst))
```

```python
#in a range,return the sum of all the palindrome numbers
def sum_of_palindrome(start, end):
    sum_pal = 0
    for i in range(start, end):
        if is_palindrome(i):
            sum_pal += i
    return sum_pal

print(sum_of_palindrome(20, 2000))
```

```python
def get_even_numbers(lst1):
    even_numbers = []
    for num in lst1:
        if num % 2 == 0:
            even_numbers.append(num)
    return even_numbers

print(get_even_numbers((121, 13331, 23221)))
```

```python
#Given a list of integers lst and an integer val, return True if all the elements in lst are factors of the val, False otherwise..
def factors(lst, val):
    for number in lst:#3
        if val % number != 0:
            return False
    return True

lst = [3, 5, 7, 2, 8]
val = 3
print(factors(lst, val))
```
