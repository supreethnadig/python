```python
def pillar_distance(n, d, w):
    return  (n - 1) * d + (n - 2) * w

print(pillar_distance(9, 3, 2))
```

```python
def finding_index(lst,value):
    for i in range(len(lst)):
        if lst[i] == value:
            return i
    return -1
```

```python

print(finding_index([1, 2, 3],3))
```

```python
def is_sum_of_two_three_digit_numbers(n):
    for i in range(100, 1000):
        if 100 <= n - i < 1000:
            return True
    return False
```

```python
print(is_sum_of_two_three_digit_numbers(1020))
```

```python
def largest_element(lst):
    lar_num == lst[0]
    for i in lst:
        if i > lar_num:
            lar_num = i
        return lar_num
```

```python
print(largest_element([1, 2, 3])
```

```python
def sum_multiples(range_1000):
    total = 0
    for num in range(range_1000):
        if num % 3 == 0 or num % 5 == 0:
            total += num
    return total

print(sum_multiples(1000))
```

```python
def sum_square_diff(n):
    sum_of_square = n * (n + 1) * (2 * n + 1) // 6
    square_of_sum = (n * (n + 1) // 2) ** 2
    diff = square_of_sum - sum_of_square
    return diff

print(sum_square_diff(100))
```

```python
def count_consonants(s):
    consonants = "bcdfghjklmnpqrstvwxyz"
    count_of_consonants = 0
    for ch in s:
        if ch in consonants:
            count_of_consonants += 1

    return count_of_consonants
```

```python
print(count_consonants("cows"))
```
